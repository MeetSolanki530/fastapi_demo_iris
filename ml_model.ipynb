```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv',usecols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',
       'Species']) #Loading Iris CSV File
```

Above we have used `usecols` to load specific columns from pandas DataFrame Not All Columns (Dataset Contains `id` column which is not required so we used this approach alternatively we load data and drop `id` columns using `df.drop()` method.)

Before split data into train and test convert Species Column using Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Species_le'] = le.fit_transform(df['Species'])
```

```python
le.classes_
```

```python
print(df['Species_le'].unique())
```

Scale Data Using `StandardScaler`

```python
from sklearn.preprocessing import StandardScaler


df['SepalLengthCm_scale'] = StandardScaler().fit_transform(df[['SepalLengthCm']])
df['SepalWidthCm_scale'] = StandardScaler().fit_transform(df[['SepalWidthCm']])
df['PetalLengthCm_scale'] = StandardScaler().fit_transform(df[['PetalLengthCm']])
df['PetalWidthCm_scale'] = StandardScaler().fit_transform(df[['PetalWidthCm']])
```

Since there is no Null Value in dataset so we can proceed for splitting the dataset into training and testing part.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[["SepalLengthCm_scale","SepalWidthCm_scale","PetalLengthCm_scale","PetalWidthCm_scale"]],df['Species_le'],random_state=20,train_size=0.8)
```

Let Us Train Classification Model On Data

```python
#Try With Randomforestclassifier an ensemble learning approach
```

```python
from sklearn.ensemble import RandomForestClassifier

re = RandomForestClassifier(criterion="gini",n_estimators=20)
```

```python
re.fit(X_train,y_train)
```

```python
y_pred  = re.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_true=y_test,y_pred=y_pred)
```

For Learning Purpose We Have Not Used Sklearn Pipeline But For API Work We Will Implement Pipeline Here.

```python
### Sklearn PipeLine
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Define preprocessing 
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
    ])
```

```python
preprocessor
```

```python
# Define pipeline: preprocessing + classifier
```

```python
pipeline = Pipeline(steps=[
    ('preprocessing',preprocessor),
    ('classifier',RandomForestClassifier(criterion="gini",n_estimators=20))
])
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']],df['Species_le'],random_state=20,train_size=0.8)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
accuracy = accuracy_score(y_true=y_test,y_pred=y_pred)
```

```python
print(f"Final Accuracy Using Sklearn Pipeline : {accuracy} Which is {accuracy * 100}% .")
```

```python
### Save Model Pipeline

import pickle

with open('Pipelinemodel.pkl','wb') as f:
    pickle.dump(pipeline,f)
```
